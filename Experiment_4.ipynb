Experiment 4: Study of Sets in Python

```
# Creation of Set
set1={"apple","banana","cherry"}
print(set1)
```

```
# Indexing not allowed
set1=("apple","banana","cherry")
print(set1[])
```

```
# Duplicates not allowed
set1={"apple","banana","cherry","apple"}
print(set1)
```

```
# The values True and 1 are considered the same value in sets,and
set1={"apple","banana","cherry",True,1,2}
print(set1)
```

```
# To check membership
set1={"apple","banana","cherry"}

print("banana" in set1)
print("Grapes" in set1)
```

```
# Once a set is created,we cannot replace its items,
# but you can add new items.
set1={"apple","banana","cherry"}

set1.add("orange")
print(set1)
```

```
# Remove items
set1={"apple","banana","cherry"}
set1.remove("banana")
print(set1)
```

```
# Set Operations
A={1,2,3,4}
B={3,4,5,6}

print("Union:",A|B)
print("Intersection:",A & B)
print("Difference1:",A-B)
print("Difference2:",B-A)
print("Symmetric Difference:",A^B)
```

```
# frozen set
x =frozenset({"apple","banana","cherry"})
print(x)
print(type(x))
```

```
# Real-time Problems on Sets
# Problem Statement 1: Unique Event Participants
list1=["Akash","Bhuvan","Chetan","Diya","Chetan","Bhuvan"]
Set_1= set(list1)
print(Set_1)
```

```
# Problem Statement 2: Three students choose elective subjects
Student1={"Chemistry","Maths","Physics","Science"}
Student2={"Hindi","Maths","English","Science"}
Student3={"Maths","Science","Biology","Science"}

common_subjects= Student1 & Student2 & Student3
print(common_subjects)
```

```
# Problem Statement 3: One set contains students in the Cricket Club
cr={"Aarti","Balraj","Ketan"}
f={"Balraj","Shiv","Raya","Aarti"}

print("Both Clubs:", cr & f)
print("Only in 1 Club:", cr^f)
```

```
# Problem Statement 4: Absent Students
present_students={"riya","amit","alok"}
all_students={"janvi","krish","nayan","ankita","riya","amit","alok"}
Absent_Students= all_students-present_students
print("Absent Students:",Absent_Students)
```

```
# Problem Statement 5: Updated Course Codes
courses={"ETC101","ETC102","ETC103"}
courses.discard("ETC102")
print("Updated Courses:", courses)
```
